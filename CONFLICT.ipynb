### Конфликт начинается с того момента, когда одна сторона решает, что притязания второй несовместимы с ее собственными. Бывают латентные(не затрагиваются базовые потребности(Маслоу - безопасность, идентиность)), открытые. Как правило избегание конфликтов ведет к краху. Обычно бесконфликтные общества умирают(излишнее миролюдие).

### _Объективные источники конфликта_ - это проблемы, которые должны быть решены или вопросы, по которым следует менять решение. Можно принять\изменить свою позицию.

### _Псевдообъективные источники конфликта(эмоциональные)_ - это эмоциональные источники, выдываемые за объективные.

### Конфликт, который эмоционально не окрашен обычно не рассматривается как конфликт. По сути конфликты имеют эмоциональный заряд.

### _Эмоциональные источники конфликта_

1. Эмоции, связанные с потребностью контролировать людей, влиять на них, добиваться желаемого социального статуса
2. Эмоции, связанные с потребностью получать одобрение со стороны других людей, переживать принадлежнось в значимой для себя группе
3. Эмоции, связанные с потребонстью справедливости, со стремлением к равенству и честности во взаимоотношениях
4. Эмоции, связанные с самоидентификацией, с потребностью в автономности, самореализации, позитивном образе собственного "Я", в утверждении собственной ценностей

### _Процесс и динамика конфликта_. Конфликт имеети кон- и де-структивные.

### _Стороны конфликта_ - это субъект социального взаимодейсвтия, находящиееся в состоянии конфликта или же явно\неявно поддерживающий конфликтующих.

### _Предмет конфликта_ - то, из-за чего возникает конфлиет

### _Образ конфликтной ситуации_ - отображение предмета конфликта в сознании субъекта конфликтного взаимодействия.

### _Ситуации и мотивы конфликта_ - внутренние побудительные силы, подталкивающие субъектов социанльных индивидов к конфликту.

### _Позиции конфликтующих сторон_ - то, о чем они завяляют друг другу в ходе конфликта или в переговорном процессе

### Конфликт = участники + объект + инцидент + конфликтная ситуация

### Индидент - пркатические дейсвтия участников в конлфиктной ситуации, которая характеризуется бескомпромиссностью посутпок и направленной на обязательное владение предметом вовлеченного интереса

### _Альтернативы модели притязаний_

1. Нет предполагаемого расхождения интересов, потому что уже известная альтернатива удовлетворяет притязаниям обеих сторон (конфликт как бы есть, но уже есть договоренности - есть альтернатива)
2. Притязания первой стороны поднимаются до такого уровня, что альтернативу, способствующую взаимному удовлетворению, подобрать невозможно
3. Притязания второй стороны достигают уровня, при котором невозможно подобрать альтернативу для взаименого удовлетворения
4. (самая ужасная) взаимоприемлимая альтервнатива недосягамема, а оставшиеся альтернативы с нулевой суммой, в которой выигрыш одной стороны означает проигрышь другой

### _Стадии конфликтной ситуации_

1. Пред
2. Сам конфликт
3. Пост
